## Make API Request

```python
import re
import time
```

```python
def remove_punctuation(input_string):
    no_punc= re.sub(r'[^\w\s]', '', input_string)
    return re.sub(r'\s+', ' ', no_punc).strip()
```

```python
# delete apostrophes and commas
remove_punctuation('empina o fuzil // power trip')
```

```python
# loop through every item and try to get responses
minidata = lnos[:]
album_urls = []
preview_urls = []
responses = []

time.sleep(61)
for i in range(len(minidata)):
    if((i+1)%20==0):
        time.sleep(1)
    # create search term
    songName = '+'.join(remove_punctuation(minidata.iloc[i]['song name']).split(' '))
    artistName = '+'.join(remove_punctuation(minidata.iloc[i]['artist']).split(' '))
    term = songName + '+' + artistName
    print(term)

    #make request
    response=requests.get('https://itunes.apple.com/search?term='+term+'&limit=1')
    print(response)


    album_url=None
    preview_url=None
    
    if response.status_code == 200:
        response_json = response.json()
        responses.append(response_json)
        if(response_json['resultCount']==1):
            album_url = response_json['results'][0]['artworkUrl100']
            preview_url = response_json['results'][0]['previewUrl']
    else:
        responses.append('No Results')

    #album url and preview url
    album_urls.append(album_url)
    preview_urls.append(preview_url)
```
